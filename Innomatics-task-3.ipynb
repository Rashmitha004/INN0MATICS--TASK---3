```python
n=int(input())
employees={}
for i in range(n):
    name=input()
    score=int(input())
    employees[name]=score
highscore=max(employees.values())
top_performers=[name for name,score in employees.items() if score==highscore]
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highscore})")
```

```python
import string
query=input().lower()
for p in string.punctuation:
    query=query.replace(p,"")
words=query.split()
frequency={}
for word in words:
    frequency[word]=frequency.get(word,0)+1
result={word:count for word,count in frequency.items() if count>1}
print(result)
```

```python
readings=list(map(int,input().split()))
valid=[]
for index,value in enumerate(readings):
    if value%2==0:
        valid.append((index, value))
print("Valid Sensor Readings:")
print(valid)
```

```python
n=int(input())
emails=[]
for i in range(n):
    emails.append(input())
domain_count={}
for email in emails:
    domain=email.split("@")[1]
    domain_count[domain]=domain_count.get(domain,0)+1
total=len(emails)
for domain,count in domain_count.items():
    percentage=(count/total)*100
    print(f"{domain}: {int(percentage)}%")
```

```python
sales=list(map(int,input().split()))
average=sum(sales)/len(sales)
for index,value in enumerate(sales):
    if value>average*1.3:
        print(f"Day {index + 1}: {value}")
```

```python
user_ids=input().split()
freq={}
for i in user_ids:
    freq[i]=freq.get(i,0)+1
for user,count in freq.items():
    if count>1:
        print(f"{user} → {count} times")
```
